# Introdução

Este projeto consiste em aplicar tecnicas de vetorização a fim de analisar suas vantagens quanto ao ganho de desempenho.
   
   Vectorização é o processo de usar registradores vetoriais, ao invés de registradores escalares, em uma tentativa de fazer o programa rodar mais rápido, ao converter um algoritimo de operar um valor por vez para operar um conjusto de valores(vetor) de uma so vez.
   
   Para tal processo a flag ftree-vectorize pode ser utilizada durante a compilação, esta possui duas flags adicionais que definem outros tipos de arquitetura do processo de vetorização como mavx e mavx2. 
   Ambas estruturas possuem maior desempenho, por possuirem um vetor maior para armazenamento, mas ao comparar o mavx com o mavx2, este se destaca neste quesito, logo tem-se que essas arquiteturas apresentadas possuem um limite, definido pelo tamanho do registrador vetorial, quanto ao ganho de eficiência no tempo de execução.
   
   Para as analízes que serão efetuadas ao decorrer deste documento, inicialmente espera-se que, para funções vetorizadas essa seja a sequência de melhor tempo de execução : mavx2, mavx , vec(ftree-vectorize) e norm(sem vetorização). Afirmação esta que sera corroborada ou refutada pelas analises a serem realizadas.

````python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import json
````

#### Load Json Files

````python
#Load files
with open('Tempo_norm.json') as data_file: 
    data_N = json.load(data_file)
    
with open('Tempo_vec.json') as data_file: 
    data_V = json.load(data_file)
    
with open('Tempo_mavx.json') as data_file: 
    data_VX = json.load(data_file)
    
with open('Tempo_mavx2.json') as data_file: 
    data_VX2 = json.load(data_file)
````

#### Codigo Assembler

O codigo assembler gerado apos a vetorização deixa de utilizar os registradores escalares utilizando registradores %xmm reverentes ao registrador vetorial.

EXEMPLO:
```
LFB0:
        vmovdqu    (%rdi), %xmm0
        movslq     16(%rdi), %rdx
        vpmovsxdq  %xmm0, %xmm1
        vpsrldq    $8, %xmm0, %xmm0
        vpmovsxdq  %xmm0, %xmm0
        vpaddq     %xmm0, %xmm1, %xmm0
        vpsrldq    $8, %xmm0, %xmm1
        vpaddq     %xmm1, %xmm0, %xmm0
        vmovd      %xmm0, %rax
        addq       %rdx, %rax
        ret
```
